```python
def reform_mat_arr(mat):
    y_arr=[]
    norm_factor =np.abs(complex(mat[0]))/complex(mat[0])
    for i,a in enumerate(mat):
        ac = complex(a)*norm_factor
        y = round(np.real(ac),4)+1j*round(np.imag(ac),4)
        y_arr.append(y)
    return y_arr

def reform_mat(mat):
    y_arr=0+0j
    
    if np.abs(complex(mat[0]))>0.1:
        norm_factor =np.abs(complex(mat[0]))/complex(mat[0])
    elif np.abs(complex(mat[1]))>0.1:
        norm_factor =np.abs(complex(mat[1]))/complex(mat[1])
    elif np.abs(complex(mat[2]))>0.1:
        norm_factor =np.abs(complex(mat[2]))/complex(mat[2])
    else:
        norm_factor =np.abs(complex(mat[3]))/complex(mat[3])
    
    for i,a in enumerate(mat):
        ac = complex(a)*norm_factor
        y = round(np.real(ac),2)+1j*round(np.imag(ac),2)
        y_arr += y*10**i
    return y_arr
```

```python
New_states=[]
for aaa in All_states:
    New_states.append( reform_mat(aaa) )
    New_states.append( reform_mat(aaa)*-1 )
    New_states.append( reform_mat(aaa)*1j )
    New_states.append( reform_mat(aaa)*-1j )
set(New_states)
```
