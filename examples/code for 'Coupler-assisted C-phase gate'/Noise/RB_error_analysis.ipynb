```python
def Lt(jump_operator,density):
    return jump_operator*density*jump_operator.dag()-0.5*jump_operator.dag()*jump_operator*density - 0.5*density*jump_operator.dag()*jump_operator
```

```python
def TimeEvolution(H,gamma,jump_operator,rho,dt):
    points=10000
    for i in range(points):
        rho = rho - 1j*(H*rho-rho*H)*dt/points + Lt(gamma*jump_operator,rho)*dt/points
        rho = rho/rho.norm()
    return rho
```
